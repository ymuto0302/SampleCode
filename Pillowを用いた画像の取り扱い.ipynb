# Pillow (PIL) を用いて画像を扱う
開発の停止した PIL (Python Image Library) からフォークされたライブラリ Pillow を用いる。

(メモ) ライブラリ名は Pillow なのに，PIL をインポートするからややこしい。

```python
from PIL import Image
from IPython.display import display

# 画像ファイルのオープン & 表示
img = Image.open('/content/flowers/flowers/flower_photos/test/daisy/10172379554_b296050f82_n.jpg')
# img.show() # 通常は show() メソッドでよいが，Colab の場合はうまくいかない
display(img) # Colab の場合，IPython.display.display を用いる
```

```python
# 画像に関する情報の取得
print("フォーマット: {}".format(img.format))
print("モード: {}".format(img.mode))
print("幅: {}".format(img.width))
print("高さ: {}".format(img.height))
print("幅 x 高さ: {}".format(img.size))
```

```python
# 画像の 90度回転
# (メモ) このテクニックはデータ拡張 (data augmentation) にて用いる
new_img = img.rotate(90)
display(new_img)
```

```python
# RGB ヒストグラム
# histogram() メソッドの出力は 256*3=768次元であり，
# 「0~255 が red, 256〜511が green, 512〜767 が blue のヒストグラムとなっている。
hist = img.histogram()
red_hist = hist[0:256]
green_hist = hist[256:256*2]
blue_hist = hist[256*2:]

# matplotlib を用いた可視化
import matplotlib.pyplot as plt
plt.plot(red_hist, c='r', label='red')
plt.plot(green_hist, c='g', label='green')
plt.plot(blue_hist, c='b', label='blue')
plt.legend()
plt.show()
```
